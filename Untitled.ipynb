```python
import scipy.io
import torch
```

```python
# 1. Load the dataset and metadata
mat_file_path = 'agus2010_cortical.mat'
print(f"Loading {mat_file_path}...")
mat_data = scipy.io.loadmat(mat_file_path, squeeze_me=True, struct_as_record=False)
```

```python
continuous_features = mat_data['features'] # Shape: [TotalFrames, Freq, Scale, Channels]
exp_config = mat_data['exp_config']
```

```python
labels = exp_config.labels                   # Shape: [200] (0=N, 1=RN, 2=RefRN)
onsets = exp_config.trial_onsets_frames - 1  # Convert to 0-indexed Python arrays
frames_per_trial = int(exp_config.frames_per_trial)
```

```python
_, Freq, Scale, Channels = continuous_features.shape
n_trials = len(labels)
print(f"Slicing {n_trials} trials ({frames_per_trial} frames per trial)...")
```

```python
# 2. Slice the continuous stream into discrete 1-second trials
trials_list = []
for i in range(n_trials):
    start_idx = int(onsets[i])
    end_idx = start_idx + frames_per_trial
   
    # Extract the 1-second segment
    trial_tensor = continuous_features[start_idx:end_idx, :, :, :]
    trials_list.append(trial_tensor)
```
